<h1> Lesson 10: Matplotlib bar charts and histograms</h1>

The simplest bar chart has two lists as inputs. One list should contain the labels for the charts, the second one the bar heights. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
ax.set_title("Very simple bar chart")
ax.bar(x,y)
plt.show()
```

In the example below two extra parameters were added :
<ul> <li> edgecolor was set to black. ('k' is black)</li> <li>color - can be one letter standard color, hex code or list of hex codes</li></ul>

```python
fig, ax = plt.subplots()
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
ax.set_title("Very simple bar chart")
ax.bar(x,y, edgecolor = "k", color = "r")
plt.show()
```

Other parameters that you may wish to change: 
<ul><li>width - width of the bar</li><li> linewidth - the width of the edge lines, </li><li>hatch - hatch pattern. Choices are : {'/', '|', '-', '+', 'x', 'o', 'O', '.', '*'}</li></ul>

```python
fig, ax = plt.subplots()
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
ax.set_title("Very simple bar chart")
ax.bar(x,y, edgecolor = "k", color = ["#4287f5", "#473045", "#314730"], width = 0.2, linewidth = 0.5, hatch = "*")
plt.show()
```

Adding the <b>error bars</b> to the bar chart. The error bar values can be stored in the list and to add them to the chart <code>yerr</code> is used for vertical error barrs and <code>xerr</code> for horizontal. To define the format (scatter marker) of the error bars the <code>fmt</code> is used.
<br>
<p>The values passed in the errorbars can be of shape N (list), or (2, N) - matrix.</p>
<ul>
<li>scalar: Symmetric +/- values for all data points.</li>
<li>shape(N,): Symmetric +/-values for each data point.</li>
<li>shape(2, N): Separate - and + values for each bar. First row contains the lower errors, the second row contains the upper errors.</li>
<li>None: No errorbar.</li></ul>

```python
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
error_bars = [1, 1.5, 0.8, 0.5]
fig, ax = plt.subplots()
ax.set_title("Very simple bar chart")
ax.bar(x,y, edgecolor = "k", color = "y", width = 0.2, linewidth = 0.5)
ax.errorbar(x, y, yerr=error_bars, fmt="D", color="r", ecolor = "k", capsize = 5)
plt.show()
```

<p>Example of the errorbars with (2, N) parameter</p>

```python
import numpy as np
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
error_bars = [[0.1, 0.3, 0.5, 1.0 ], [0.2, 0.4, 0.1, 0.5]]
fig, ax = plt.subplots()
ax.set_title("Very simple bar chart")
ax.bar(x,y, edgecolor = "k", color = "y", width = 0.2, linewidth = 0.5)
ax.errorbar(x, y, yerr=error_bars, fmt="D", color="r", ecolor = "k", capsize = 5)
plt.show()
```

How to acess the specific bar in the bar chart. You can do it by iterating through them: 


```python
fig, ax = plt.subplots()
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]
ax.set_title("Very simple bar chart")
bar_chart = ax.bar(x,y, edgecolor = "k", color = "#473045", width = 0.2, linewidth = 0.5, hatch = "O")
for num, bar in enumerate(bar_chart):
    if num == 2:
        bar.set_color("r")
plt.show()
```

<h3> Grouped bar charts</h3>
<p> There are different ways to draw the <b>grouped bar chart</b></p>
<p> 1. First, the most manual way to do it: </p>

```python
fig, ax = plt.subplots()
# data 
x = np.array([1, 2, 3, 4, 5])
y1 = [34, 56, 12, 89, 67] 
y2 = [12, 56, 78, 45, 90] 
y3 = [14, 23, 45, 25, 89] 
width = 0.1
# plot data in grouped manner of bar type 
ax.bar(x-0.2, y1, width, color='r', edgecolor = "k") 
ax.bar(x, y2, width, color='b', edgecolor = "k") 
ax.bar(x+0.2, y3, width, color='y', edgecolor = "k") 
ax.set_xticks(x, ['A', 'B', 'C', 'D', 'E']) 
ax.set_xlabel("Groups") 
ax.set_ylabel("Scores") 
ax.legend(["Measurement 1", "Measurement 2", "Measurement 3"]) 
plt.show() 
```

Why we need an <b>np array</b> here? 
Because if I would use a list three different lists would be needed. While the <code>np.array</code> allows to generate three sets of numbers just by substracting the width. 

```python
my_array = np.array([1, 2, 3, 4, 5])
width_example = 0.3
result = my_array - width_example
print("Original array")
print(my_array)
print("Array after substraction")
print(result)
```

2. Automatic way to create the bar chart using the <b>pandas dataframe </b>

```python
df_sample = pd.DataFrame([['Group1', 10, 20, 10, 30], ['Group2', 20, 25, 15, 25], ['Group3', 12, 15, 19, 6], 
                   ['Group4', 10, 29, 13, 19]], 
                  columns=['Group nr', 'Measurement 1', 'Measurement 2', 'Measurement 3', 'Measurement 4']) 
df_sample
```

```python
df_sample.plot(x='Group nr', 
        kind='bar', 
        stacked=False, 
        title='Grouped Bar Graph with dataframe', edgecolor = "k")
plt.show()
```

<h3> Histograms</h3>
<p> Histogram a chart that plots the distribution of a numeric variable's values as a series of bars.</p>
To give you an example of the histogram lets's plot something from our <code> metabric dataset</code>

```python
df_hist = pd.read_csv("METABRIC_RNA_Mutation.csv")
```

```python
data_hist = df_hist["age_at_diagnosis"].to_list()
```

The histogram parameters
<ul><li>x - compulsary parameter, the list or array containing the data. </li>
<li>bins - how many bars you will have, default is <b>10</b></li><li>orientation: horizontal or vertical</li> <li> rwidth - 0 to 1 fraction of space that band occupies relative to the space allocated to the band</li></ul>

```python
fig, ax = plt.subplots()
ax.hist(x = data_hist, bins = 15, orientation = 'vertical', color = "#f2d583", edgecolor = "k", rwidth = 0.8)
ax.set_xlabel("Age at diagnosis")
ax.set_ylabel("Number of patients")
ax.set_yscale("log")   
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(x = data_hist, bins = 15, orientation = 'horizontal', color = "#f2d583", edgecolor = "k", rwidth = 0.8, histtype = 'barstacked')
ax.set_xlabel("Age at diagnosis")
ax.set_ylabel("Number of patients")
plt.show()
```
