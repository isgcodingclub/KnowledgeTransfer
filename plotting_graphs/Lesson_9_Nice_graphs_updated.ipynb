<h3> Adding the grid </h3>
Grid on its own has many parameters. You can read about all of them at:
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html I will list only some of them: <ul><li> color - defines colour of the grid. <ul> <li> You can put one letter colour symbols like "r" for red, <b>"y"</b> for yellow. The notations are: {'b', 'g', 'r', 'c', 'm', 'y', 'k', 'w'} for blue, green, red, cyan, magenta, yellow, black, and white </li> <li> you can put the hex colour codes e.g "#3b1fec" </li></ul> <li> alpha - transparency of the gridlines 0 - invisible, 1 - opaque </li> <li> linestyle defines the style of the line Available options: '-', '--', '-.', ':', '', </li> <li> linewidth - width of the line.  </li></ul>

````python
fig, axes = plt.subplots()
axes.set_title("title of graph")
axes.grid(color='#3b1fec', linestyle=':', linewidth=0.7, alpha = 0.7)
````

<b> How to change the grid size? </b>

````python
import numpy as np
fig, ax = plt.subplots()
major_ticks = np.arange(0, 1.2, 0.2) # two numpy arrays to define the major and minor gridlines
minor_ticks = np.arange(0, 1.1, 0.1)
ax.set_xticks(major_ticks) # setting the major gridlines for x axes
ax.set_xticks(minor_ticks, minor=True) # setting the major gridlines for x axes
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)
ax.grid(which='both')
ax.grid(which='minor', alpha=0.2, color = "b", linestyle = "-.")
ax.grid(which='major', color = "k", linestyle = "--", alpha = 0.8)
````

<h4>Customising the legend</h4>
<p>In the example below there are 3 lines in the plot, but only two are passed in the legend as the <code>handles</code></p>

````python
fig, ax = plt.subplots()
ax.set_title("OD versus time")
ax.set_ylabel("Y axis")
ax.set_xlabel("X axis")
ax.set_facecolor("#f5ebd3")
ax.grid(linestyle = ":", linewidth = 0.5, color = "k")
time = [0, 20, 40, 60, 80]
OD = [0.0140, 0.0326, 0.0684, 0.1372, 0.2794]
OD1 = [0.0594, 0.1125, 0.2173, 0.4245, 0.8376]
OD2 = [0.04, 0.07, 0.14, 0.28, 0.56]
red_ln, = ax.plot(time, OD, color = "r", linewidth = 2, linestyle = "--") # you absolutely need the , if you want to add lines to the legend handles
blue_ln, = ax.plot(time, OD1, color = "b", linewidth = 2, linestyle = "--")
yell_ln = ax.plot(time, OD2, color = "y", linewidth = 2, linestyle = "--")
ax.legend(handles = [red_ln, blue_ln], labels = ["First line", "Second line"]) # manually assign labels
plt.show()
````

Changing the position of the legend.
The options are:
<ul><li>upper right</li> <li> upper left</li> <li>lower left</li><li>lower right</li> <li>right</li> <li>center left</li> <li>center right</li><li>lower center</li><li>upper center</li> <li>center </li> <li> best - places legend further away from other elements</li></ul>

````python
fig, ax = plt.subplots()
ax.set_title("OD versus time")
ax.set_ylabel("Y axis")
ax.set_xlabel("X axis")
ax.set_facecolor("#f5ebd3")
ax.grid(linestyle = ":", linewidth = 0.5, color = "k")
time = [0, 20, 40, 60, 80]
OD = [0.0140, 0.0326, 0.0684, 0.1372, 0.2794]
OD1 = [0.0594, 0.1125, 0.2173, 0.4245, 0.8376]
ax.plot(time, OD, color = "r", linewidth = 2, linestyle = "--")
ax.plot(time, OD1, color = "b", linewidth = 2, linestyle = "--")
ax.legend(labels = ["First line", "Second line"], loc = "center") # assign the location of the legend
plt.show()
````

Another way how to set up the position of the legend is <code>bbox_to_anchor</code>.
<p> Accepts either 2-tuple, or 4-tuple of floats </p>
<p><b>The 4 parameter <code>bbox_to_anchor</code></b></p>
<p> If you provide the 4-tuple input the parameters are (x, y, width, height) x and y are coordinates of the <b>left corner</b> of the bounding box and width and height are the wdth and height of the box. </p>
<p>The <code>loc</code> parameter then specifies which corner of the legend attaches to which corner of that bounding box</p>
<p>If you do not specify <code>loc</code> parameter, it uses default value <code>loc = 'best'</code></p>

````python
fig, ax = plt.subplots()
ax.set_title("OD versus time")
ax.set_ylabel("Y axis")
ax.set_xlabel("X axis")
ax.set_facecolor("#f5ebd3")
ax.grid(linestyle = ":", linewidth = 0.5, color = "k")
time = [0, 20, 40, 60, 80]
OD = [0.0140, 0.0326, 0.0684, 0.1372, 0.2794]
OD1 = [0.0594, 0.1125, 0.2173, 0.4245, 0.8376]
ax.plot(time, OD, color = "r", linewidth = 2, linestyle = "--")
ax.plot(time, OD1, color = "b", linewidth = 2, linestyle = "--")
ax.legend(labels = ["First line", "Second line"], bbox_to_anchor=(-0.2, 0.5, 0.8, 0.5), loc = "upper right") # assign the location of the legend
plt.show()
````

<p><b>The 2 parameter <code>bbox_to_anchor</code></b></p>
<p>If you are using just the (x, y) paraemeter - 2-tuple, this just specifies where the corner of the legend will be located, the <code>loc</code> parameter specifies which corner will be located at x, y</p>



````python
fig, ax = plt.subplots()
ax.set_title("OD versus time")
ax.set_ylabel("Y axis")
ax.set_xlabel("X axis")
ax.set_facecolor("#f5ebd3")
ax.grid(linestyle = ":", linewidth = 0.5, color = "k")
time = [0, 20, 40, 60, 80]
OD = [0.0140, 0.0326, 0.0684, 0.1372, 0.2794]
OD1 = [0.0594, 0.1125, 0.2173, 0.4245, 0.8376]
ax.plot(time, OD, color = "r", linewidth = 2, linestyle = "--")
ax.plot(time, OD1, color = "b", linewidth = 2, linestyle = "--")
ax.legend(labels = ["First line", "Second line"], bbox_to_anchor=(0.5, 0.5), loc = "upper right") # assign the location of the legend
plt.show()
````

<p> Other parameters of the legend that can be changed: </p>
<ul> <li> ncols (number of columns) </li>
<li>facecolor: - the colour of the legend background(you put the hex colour code)</li>edgecolor: the legend frame colour,  (you put the hex colour code)<li>shadow (use True or false)</li><li>fontsize, the size of the text</li>
<li>Fancybox : True or False - specifies whether the legend has a round coreners or not. True is default and corresponds round corners, false is for sharp corners</li></ul>

````python
fig, ax = plt.subplots()
ax.set_title("OD versus time")
ax.set_ylabel("Y axis")
ax.set_xlabel("X axis")
ax.set_facecolor("#f5ebd3")
ax.grid(linestyle = ":", linewidth = 0.5, color = "k")
time = [0, 20, 40, 60, 80]
OD = [0.0140, 0.0326, 0.0684, 0.1372, 0.2794]
OD1 = [0.0594, 0.1125, 0.2173, 0.4245, 0.8376]
ax.plot(time, OD, color = "r", linewidth = 2, linestyle = "--")
ax.plot(time, OD1, color = "b", linewidth = 2, linestyle = "--")
ax.legend(labels = ["First line", "Second line"], ncols = 2, edgecolor = "#b71b14",
          shadow = True, facecolor = "#d2c3c2", fontsize= 15, fancybox = False, title="My legend", title_fontsize = 15) # assign the location of the legend
plt.show()
````

<h3>Making the scatter graphs in matplotlib</h3>
There is a <b>marker</b> parameter.
Here is the link to marker options: <a href = "https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers">Marker options </a>
<p><b>s</b> - refers to the marker size</p>
<p><b>color</b> - marker color</p>
<p> Not all markers have the <b> edgecolor</b> option </p>


````python
time = [0, 20, 40, 60, 80]
OD = [0.0140, 0.0326, 0.0684, 0.1372, 0.2794]
fig, ax = plt.subplots()
ax.set_title("The scatter graph")
ax.scatter(x = time, y = OD, marker = "D", color = 'r', edgecolors = 'k', s = 40)
ax.legend(labels = ["OD measurements"])
plt.show()
````
