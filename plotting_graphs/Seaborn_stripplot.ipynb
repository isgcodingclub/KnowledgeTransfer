<h3>Categorical stripplots</h3>
<p>Stripplots require additional package named seaborn </p>
<p>Can be installed with <code>!pip install seaborn</code></p>
<p>In this example the Metabric dataset is used. Can be downloaded from here <a href = "https://github.com/isgcodingclub/KnowledgeTransfer/blob/main/datasets/METABRIC_RNA_Mutation.csv">Metabric DNA mutation dataset</a></p>

```python
import seaborn as sns # importing seaborn
import pandas as pd # importing pandas
import matplotlib.pyplot as plt # importing the metabric dataset
```

```python
df = pd.read_csv("METABRIC_RNA_Mutation.csv")
df.head()
```

<p>Below is the example of the stripplot plot, lets have a look at parameters</p>
<ul><li>data - a pandas dataframe</li>
<li>x - the name of the column that will contain x values</li>
<li>y - the name of the the column that will contain the y values</li>
<li>ax - should be set equla to the axes object obtained as a result of <code>fig, ax = plt.subplots()</code></li>
<li>edgecolor - the color of the marker countours, this will only be visible if <code>linewidth</code>parameter is > 0</li>
<li>hue - the column that defines the category, each category will be colored in diferent colour</li>
<li>marker - the shape of the marker</li>
<li>palette - string, list, dict, or matplotlib.colors.Colormap</li>
</ul>

```python
fig, axes = plt.subplots()
sns.stripplot(data = df, x = "type_of_breast_surgery", y = "age_at_diagnosis", 
              ax = axes, edgecolor = "k", 
              linewidth = 0.5, hue = "cancer_type_detailed", marker = "D", 
              palette = ["#5af16c", "#b2112c", "#7e9ebd", "#f3be82", "#1423a3", "#FFEF00"])
plt.legend(loc = "upper left", bbox_to_anchor = (1, 1.1))
plt.show()
```
