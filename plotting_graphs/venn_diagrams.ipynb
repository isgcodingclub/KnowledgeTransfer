<p>Venn package requires the data to be dictionary of sets</p>

```python
from venn import venn, pseudovenn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read the csv using pandas 
df = pd.read_csv("s__aa-R2_joined.fastq_for_machine.csv")
CDRs = set(df["Unique_CDR_combination"].to_list())

df1 = pd.read_csv("s__aa-R3_joined.fastq_for_machine.csv")
CDRs1 = set(df1["Unique_CDR_combination"].to_list())

# convert to sets and 
dictionary_of_sets = {"aa-R2": CDRs, "aa-R3": CDRs1}
lst_colormaps = ['PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu', 'RdYlBu',
                      'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic',
                      'berlin', 'managua', 'vanimo']

# plotting the graphs with the different colormaps. 
for a in lst_colormaps:
    venn(dictionary_of_sets, cmap = a)
    plt.show()
```

<p>If you have 6 sets to compare, the venn pacakge can plot <b>pseudovenn</b> diagram</p>
<p>This diagram does not show all the comparisons, but is more visually appealing than the venn diagram for 6 sets</p>

```python
from venn import pseudovenn
import random
```

```python
# In this example I use random.sample to make the sets by selecting the subsample from amino acid list (list_to_select_from)
# then I caonvert the each selection to set and add it to dictionary
list_to_select_from = ["A", "C", "D", "E", "F", "G", "H", "I", "K", "L", "M", "N", "P", "Q", "R", "S", "T", "V", "W", "Y"]
dictionary_of_sets_ready = {}
for i in range(1, 7):
    set_to_add = set(random.sample(list_to_select_from, 8))
    dictionary_of_sets_ready[i] = set_to_add
# plotting the diagram.
pseudovenn(dictionary_of_sets_ready, cmap = "RdBu")

    
```

<h3>The matplotlib-venn package</h3>

<p>This package does not require the dictionary of sets. You just need to have the numbers in each set</p>
<p>Can do two or three circle venn diagrams</p>

```python
from matplotlib_venn import venn2
A_set_only = 10
common_elements = 5
B_set_only = 7
v = venn2(subsets=(A_set_only, B_set_only, common_elements), set_labels=("A", "B"), set_colors=('mediumaquamarine', 'slategrey'), alpha=0.75)
# this code makes the lines around the circles. 
for patch in v.patches:
    if patch:  # skip None patches
        patch.set_edgecolor('black')
        patch.set_linewidth(0.5)
plt.show()
```

```python
from matplotlib_venn import venn3
A_set_only = 10
common_elements = 5
B_set_only = 7
C_set_only = 4
A_B = 2
A_C = 3
B_C = 6
# this is how the elements within the subsets parameter needs to be places in order.
v = venn3(subsets=(A_set_only, B_set_only, A_B, C_set_only, A_C, B_C, common_elements), set_labels=("A", "B", "C"), 
          set_colors=('mediumaquamarine', 'slategrey', 'magenta'), alpha=0.75)

# this code makes the lines around the circles. 
for patch in v.patches:
    if patch:  # skip None patches
        patch.set_edgecolor('black')
        patch.set_linewidth(0.5)
plt.show()
```
