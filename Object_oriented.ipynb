<p>The example above is not particularly useful, because you can not really customise the method. Additionally, when ostrich object is created it does not have any properties.</p>
<p>There is a way to assign the properties to the object, at the time when it is created, the class needs to have the <code>def__init__(self)</code> method</p>
<p>First, the <code>__init__</code> method is executed when object is created.</p>
<p>Second, our <code>__init__</code> method has arguments, so in order to create the <code>Ostrich</code> object we will need to provide the <code>color_input</code> and <code>height_input</code></p>

````python
class Ostrich:
    def __init__(self, color_input, height_input):
        self.color = color_input 
        self.height = height_input
        print("We created this Ostrich")
    
    def print_name(self):
        print("Patrich")
````

````python
# This is to illustate that the __init__ is executed when object is created
yellow_ostrich = Ostrich("yellow", 2)
````

<p>So despite the fact that we haven't called the <code>__init__</code> method anywhere, it is executed when we create the object, and that's why the line is printed.</p>

<h3>Attributes defined within the <code>__init__</code> method and the parameter <code>self</code></h3>

<p>When we created the <code>Ostrich</code> class we put some parameters in its <code>__init__</code> method. This is how the class accepts the properties of the the Ostrich object.</p>
<p>But these properties need to be saved within the object. That's why there are things called <b>attributes</b>, which are similar to variables, but they are specific to the object.</p>
<p>Below is the example how do you can assign the attribute and then access it.</p>

````python
class Ostrich:
    def __init__(self, color_input, height_input):
        self.color = color_input  # This is an attribute
        self.height = height_input # This is another attribute
    
    def print_name(self):
        print("Patrich")

yellow_ostrich = Ostrich("yellow", 2)
print(yellow_ostrich.color) # printing the attribute
red_ostrich = Ostrich("red", 2)
print(red_ostrich.color)
````

<p>What does the parameter <code>self</code> does?</p><p> Shortly, it ensures that the attributes that are provided when object is created only apply to that particular object.</p>
<p>There is also a type of the attribute called a <b>class attribute</b> that will be stared by all ostrich objects and is defined outside the <code>def__init__</code> method</p>

````python
class Ostrich:
    number_legs = 2
    def __init__(self, color_input, height_input):
        self.color = color_input  # This is an attribute
        self.height = height_input # This is another attribute
    
    def print_name(self):
        print("Patrich")
blue_ostrich = Ostrich("blue", 2)
print(blue_ostrich.number_legs)

white_ostrich = Ostrich("white", 2)
print(white_ostrich.number_legs)

# It is a good practise to acess the class attribute not by specific object name, but by class name
print(Ostrich.number_legs)
````

````python
class Ostrich:
    # This counter is going to count how many ostriches we created
    ostrich_number = 0 # class attribute

    def __init__(self, color_input, height_input):
        self.color = color_input  # This is an attribute
        self.height = height_input # This is another attribute
        Ostrich.ostrich_number += 1 # updating the ostrich number.

first_ostrich = Ostrich("brown", 1)
print(first_ostrich.ostrich_number)
second_ostrich = Ostrich("white", 1.5)
print(first_ostrich.ostrich_number)
````

<p>We can create the class that accepts the DNA sequence and creates the compliment or reverse compliment of it.</p>

````python
compliment_dict = {"A":"T", "T":"A", "C":"G", "G":"C"}
class DNA:
    def __init__(self, sequence, transformation):
        self.sequence = sequence
        self.transformation = transformation if transformation in ("compliment", "reverse_compliment") else "compliment"

    def transform(self):
        if self.transformation == "compliment":
            result = "".join([compliment_dict.get(a, "N") for a in self.sequence])
        else:
            result = "".join([compliment_dict.get(a, "N") for a in self.sequence[::-1]])
        return result
my_DNA = DNA("ATCGCACTCAAAGCCGGGAC", "reverse_compliment")
print(my_DNA.transform())
````

<h3>Inheritance</h3>

<p>The class inheritance allows the <b>child class</b> to inherit all the attributes and the methods of the parent class.</p>
<p>This allows to reuse the code.</p>

````python
class Bird:
    def __init__(self, name):
        self.name = name

    def move(self):
        print(f"{self.name} moved")


class Ostrich(Bird):
    pass

ostrich1 = Ostrich("Patrick")
ostrich1.move()
print(ostrich1.name)
````

<p>So, what happened in the example above was that <code>class Ostrich</code> inherited the <code>move</code> method from the parent <code>class Bird</code>, same happened with the <code>self.name</code> attribute.</p>

<p>Not only child classes can inherit the methods from the parent classes, but they also can have additional methods.</p>

````python
class Bird:
    def __init__(self, name):
        self.name = name

    def move(self):
        print(f"{self.name} moved")


class Ostrich(Bird):
    def run_away(self):
        print(f"{self.name} ran away from farm")

class Pinguin(Bird):
    def swim(self):
        print(f"{self.name} swam")

ostrich_2 = Ostrich("Ozzy")
ostrich_2.move()
ostrich_2.run_away()

pingquin1 = Pinguin("Rainbow")
pingquin1.swim()
pingquin1.move()
````

<p>What happens when the child class has the same method as parent class. Spoiler. It gets overwritten</p>

````python
class Bird:
    def __init__(self, name):
        self.name = name

    def move(self):
        print(f"{self.name} moved")


class Ostrich(Bird):
    def move(self):
        print(f"{self.name} ran away from farm")

ostrich3 = Ostrich("Nugget")
ostrich3.move()
````

<p>What about the <code>def__init__</code> method?</p> <p>Child classes inherit the <code>def__init__</code> method from parent class automatically.</p>
<p>They can have their own <code>def__init__</code> method, but then it overwrites the parent class <code>def__init__</code></p>
<p>If you want the child class have its own <code>def__init__</code> method, and have additional necessary parameters that will be requested when object is created you need to use <code>super().__init__</code> in addition to the <code>def__init__</code></p>
<p>You put all the parameters including the additional ones in the <code>def __init__</code> method.</p>
<p>You put the ones you want to inherit from the parent class into the <code>super().__init__()</code> method.</p>

````python
class Bird:
    def __init__(self, name):
        self.name = name # lets create the class bird that will have one attribute name
    def move(self):
        print(f"{self.name} moved")

class Colibri(Bird):
    def __init__(self, name, size): # This method will specify all the atrivutres that che child class has
        super().__init__(name) # this will ensure that it still saves the name, even there is no self.name = name
        self.size = size

    def describe_colibri(self):
        print(f"The name of this colibri is {self.name}, \n its size is {self.size}")


colibri1 = Colibri("Jasper", "2cm")
colibri1.describe_colibri()
````

<h3>Multiple inheritance</h3>

<p>The child class can inherit from multiple parent classes.</p>

````python
class Molecule:
    def __init__(self, name):
        self.name = name
        self.orientation = 0
        
    def rotate(self):
        self.orientation += 90 if self.orientation < 360 else 0
        print(f"Orientation of {self.name} is {self.orientation}")
                  
class Polar_Amino_acid(Molecule):
    
    def ionise_side_chain(self):
        print(f"{self.name} got ionised")

class Aromatic_amino_acid(Molecule):
    def write_ring(self):
        print(f"{self.name} has a benzene ring") 
    

class Tyrosine(Polar_Amino_acid, Aromatic_amino_acid):
    pass
    
# in this example the Tyrosine class is     
Tyrosine1 = Tyrosine(True)
Tyrosine1.ionise_side_chain()
Tyrosine1.write_ring()
````

<p>We can create a protein class that will accept the DNA sequence and turn it to the protein</p>

````python
codon_table = {
    "TTT": "F", "TTC": "F", "TTA": "L", "TTG": "L",
    "TCT": "S", "TCC": "S", "TCA": "S", "TCG": "S", "TAT": "Y", "TAC": "Y",
    "TAA": "*", "TAG": "*", "TGT": "C", "TGC": "C", "TGA": "*", "TGG": "W", "CTT": "L", "CTC": "L",
    "CTA": "L", "CTG": "L", "CCT": "P", "CCC": "P", "CCA": "P", "CCG": "P", "CAT": "H", "CAC": "H", "CAA": "Q", "CAG": "Q",
    "CGT": "R", "CGC": "R", "CGA": "R", "CGG": "R", "ATT": "I", "ATC": "I", "ATA": "I",
    "ATG": "M", "ACT": "T", "ACC": "T", "ACA": "T", "ACG": "T", "AAT": "N", "AAC": "N", "AAA": "K", "AAG": "K",
    "AGT": "S", "AGC": "S", "AGA": "R", "AGG": "R", "GTT": "V", "GTC": "V", "GTA": "V", "GTG": "V",
    "GCT": "A", "GCC": "A", "GCA": "A", "GCG": "A", "GAT": "D", "GAC": "D", "GAA": "E", "GAG": "E",
    "GGT": "G", "GGC": "G", "GGA": "G", "GGG": "G"}

compliment_dict = {"A":"T", "T":"A", "C":"G", "G":"C"}

class DNA:
    def __init__(self, sequence, frame):
        # we can do that although it comes with caveats, because somebody might try to put number in the sdequence slot
        self.sequence = sequence
        self.frame = frame if frame in (1, 2, 3, -1, -2, -3) else 1
    
    def translate(self):
        if self.frame > 0:
            start = self.frame - 1
            real_sequence = self.sequence[start:]
        else:
            # the reverse frames need to be first the reverse compliment of them
            start = abs(self.frame) - 1
            reverse_comp = [compliment_dict.get(a, "N") for a in self.sequence[::-1]]
            reverse_string = ""
            for u in reverse_comp:
                reverse_string += u
            real_sequence = reverse_string[start:]

        # cutting the sequence into the ngrams of 3
        cut_sequence = [real_sequence[i:i+3] for i in range(0, len(real_sequence) - 2, 3)]
        translated = "".join([codon_table.get(codon, "X") for codon in cut_sequence])
        return translated


seq = DNA("ATCGCACTCAAAGCCGGGAC", -2)
ready = seq.translate()
print(ready)
            
        
````
